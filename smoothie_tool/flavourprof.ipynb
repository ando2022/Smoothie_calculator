```python
import openai
import pandas as pd
import time

# Make sure your API key is set
api_key = "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"  # Replace or use environment variable

client = openai.OpenAI(api_key=api_key)

# === Test Single Ingredient ===
test_response = client.chat.completions.create(
    model="gpt-4",
    messages=[
        {"role": "system", "content": "You are a culinary expert."},
        {"role": "user", "content": "What is the dominant flavor of spirulina? Choose one from: sweet, sour, bitter, salty, umami, earthy, spicy, nutty, creamy, fruity, bland."}
    ],
    temperature=0
)

test_flavor = test_response.choices[0].message.content.strip().lower()
print("🧪 Test flavor for spirulina:", test_flavor)

# === Load Data ===
full_df = pd.read_csv("/Users/anastasiiadobson/Library/CloudStorage/Dropbox/WORK/BOOTCAMP/SMOOTHIES/data/processed/full_df.csv")
print("✅ Loaded full_df:", full_df.shape)

# === Allowed Flavors ===
ALLOWED_FLAVORS = [
    "sweet", "sour", "bitter", "salty", "umami",
    "earthy", "spicy", "nutty", "creamy", "fruity", "bland"
]

# === Extract Unique Ingredients ===
unique_ingredients = full_df["Ingredient"].dropna().unique().tolist()
flavor_profiles = {}

# === Get Flavor Profiles from GPT ===
for ing in unique_ingredients:
    prompt = (
        f"What is the **dominant flavor** of the ingredient '{ing}'? "
        f"Choose only one word from this list: {', '.join(ALLOWED_FLAVORS)}."
    )
    
    try:
        response = client.chat.completions.create(
            model="gpt-4",
            messages=[
                {"role": "system", "content": "You are a culinary expert."},
                {"role": "user", "content": prompt}
            ],
            temperature=0
        )

        flavor = response.choices[0].message.content.strip().lower()

        if flavor not in ALLOWED_FLAVORS:
            print(f"⚠️ Unexpected response for {ing}: '{flavor}' → defaulting to 'bland'")
            flavor = "bland"

        flavor_profiles[ing] = flavor
        print(f"✅ {ing}: {flavor}")
        time.sleep(1.2)

    except Exception as e:
        print(f"❌ Error with {ing}: {e}")
        flavor_profiles[ing] = "bland"

# === Add to DataFrame and Save ===
full_df["FlavorProfile"] = full_df["Ingredient"].map(flavor_profiles)
full_df.to_csv("full_df_with_flavors.csv", index=False)
print("✅ Done. Saved to 'full_df_with_flavors.csv'")
```
